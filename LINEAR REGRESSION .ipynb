```python
# TRAINING THE MODEL
```

```python
dataset.head()
# there are 13 independent features
```

```python
## Independent & Dependent Features
# to see independent features we can use drop or iloc

X=dataset.iloc[:, :-1] # [:] means all the rows and [:-1] means except the last column
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
## Stanardize or Feature scaling the Dataset
### standardscaler centralize the data and their mean will be zero and their inly the distribution will happen
## in normalization it gives the range between 0 to1, 1to2 etc
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
# We use fit_transform in train data and that transformed data we use for testing thats why we use transform in test to avoiv data leakage
```

```python
X_train=scaler.fit_transform(X_train) # we aply z score formula in fit_transform and we change the data
```

```python
X_test=scaler.transform(X_test)
```

```python
# why do we use transform in X_test not fit_transform
## to avoid Data leackage
```

```python
X_train
```

```python
X_test
```

# Model TRaining

```python
# is this is slr problem or mlr problem
## always see the documentation of scikitlearn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train, y_train) # X_train is independent feature and y_train is deendent feature
```

```python
# we used fit_transform in standardscalar bcz we needed to apply the formula on entire variables and change the range
# fit means just train and fit_transform means the set and change the data
# In fit we just ally the formula not change the data and aim is to find the best fit line in short in fit we are trying to find the parameters
```

```python
## print the coefficient and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# Prediction for the test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
### Assumption of Linear Regression

# If you are getting linear assumption that means the model is good
```

```python
plt.scatter(y_test, reg_pred)
plt.xlabel("Test Truth Data")
plt.ylabel("Test Predicted Data")
```

```python
## Residual it means the difference between the test and predicted points
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
# Whenever we try to plot distance plot for the residuals it will approximately coming as normal distribution we get skewed little in right side
```

### Assumptions to check the model is coming good or bad
1. assumptions wrt the test and predicted data { it is linearly distributed from plot i.e. when x is increasing y is increasing}
2. Errors

```python
# Scatter plot with  prediction  and residuals
# Uniform distribution ( there is no pattern in that) homosidacity
plt.scatter(reg_pred,residuals)
```

```python
## Performance Matrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```
